```python
import pandas as pd
import numpy as np

# dataset
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana", None],
    "idade": [23, None, 35, 23, 29],
    "cidade": ["Maputo", "Beira", "maputo", "MAPUTO", "Nampula"],
    "salario": ["50.000", "60.000", "None", "50.000", "70.000"]
}

df = pd.DataFrame(dados)



print("\n- Dataset Inicial -")
print(df)

# Salvar como csv
df.to_csv("dados.csv", index=False)
print("\ncvs criado com sucesso.")

# Exploração inicial
print("\n- Primeiras linhas -")
print(df.head())

print("\n- Informacoes do Dataset -")
print(df.info())

print("\n- Dimensao -")
print(df.shape)

print("\n- Nomes das colunas -")
print(df.columns)

# Valores nulos
print("\n- Valores nulos (true = nulo) -")
print(df.isnull())

#limpeza de dados
df["salario"] = df["salario"].replace("None", np.nan)

df["salario"] = df["salario"].str.replace(".", "", regex=False)
df["salario"] = df["salario"].astype(float)

df["cidade"] = df["cidade"].str.lower()

print("\n- Nulos apos limpeza -")
print(df.isnull().sum())

#Inserir valores nulos
df["nome"] = df["nome"].fillna("Desconhecido")

df["idade"] = df["idade"].fillna(df["idade"].mean())
```
