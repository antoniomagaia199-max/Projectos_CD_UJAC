```python
import pandas as pd
dados = { "horas de estudo": [3, 4, 5, 2, 6, 7, 8],
         "notas": [13, 14, 15, 10, 16, 17, 18]
    
}

df = pd.DataFrame(dados)
```

```python
# Calcular correlacao
correlacao = df["horas de estudo"].corr(df["notas"])
print(f"correlacao entre horas de estudo e notas sao: {correlacao:.2f}")
```
